# Data Preprocessing

```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy
from sklearn import preprocessing

df = pd.read_csv('data.csv')
```

## Handling Missing Values

```python
# Get whether there are N/A data
df.isnull().sum()
```

```python
# Fill N/A with a value
df = df.fillna('empty')

# Drop N/A rows
df = df.dropna()

# Fill with previous cell
df = df.fillna(method='bfill')

# Fill with next cell
df = df.fillna(method='ffill')

# Fill with approproate value
df = df.fillna({'age': 10, 'sex': 1, 'cp': 1, 'trestbps': 130})

df.count()
```

## Removing Outliers

### Using Z-Score

```python
sns.distplot(df['thalach'])
```

```python
# Remove too far away points using z-score
def remove_outliers_by_z_score(dataset, column, threshold):
    # Get absolute z-scores of the series values
    z_score = scipy.stats.zscore(dataset[column])
    abs_z_score = np.abs(z_score)

    # Return rows where absolute z-score is less than threshold
    return dataset[abs_z_score < threshold]
```

```python
or_df = remove_outliers_by_z_score(df, 'thalach', 3)
sns.distplot(or_df['thalach'])
```

### Using Quantile

```python
# Remove too far away points using quantile
def remove_outliers_by_quantile(dataset, column, threshold):
    assert(0 <= threshold <= 1)
    
    q = dataset[column].quantile(threshold)
    return dataset[dataset[column] < q]
```

```python
q_df = remove_outliers_by_quantile(df, 'thalach', 0.8)
sns.distplot(q_df['thalach'])
```

## Encoding Categorical Variables

```python
df.head()
```

```python
# Encodes string/object values to integers (eg: sex-male/female into 0/1)
def label_encode(dataset, column):
    dataset_copy = dataset.copy()
    
    label_encoder = preprocessing.LabelEncoder()
    dataset_copy[column] = label_encoder.fit_transform(dataset[column])
    
    return dataset_copy
```

```python
e_df = label_encode(df, 'sex')
e_df.head()
```

## Data Normalization/Rescaling

- When our data is comprised of attributes with varying scales, many machine learning algorithms can benefit from rescaling the attributes to all have the same scale.
- This is useful for optimization algorithms in used in the core of machine learning algorithms like gradient descent.
- It is also useful for algorithms that weight inputs like regression and neural networks and algorithms that use distance measures like K-Nearest Neighbors.

```python
# Scale a series value to 0-1 range (feature_range)
def normalize(dataset, column, feature_range=(0, 1)):
    # Copy dataset and extract the series to binarize
    dataset_copy = dataset.copy()
    X = df[[column]]
    
    min_max_scaler = preprocessing.MinMaxScaler(feature_range=feature_range)
    df_scaled = min_max_scaler.fit_transform(X)
    
    dataset_copy[column] = pd.Series(df_scaled.reshape(-1))
    return dataset_copy
```

```python
n_df = normalize(df, 'chol')
n_df.head()
```

## Binarize Data/Threshold data

- We can transform our data using a binary threshold. **All values above the threshold are marked 1 and all equal to or below are marked as 0**.
- This is called binarizing your data or threshold your data. It can be useful when you have probabilities that you want to make crisp values. It is also useful when feature engineering and you want to add new features that indicate something meaningful.

```python
# Applies a threshold to a column
def binarize(dataset, column, threshold):
    # Copy dataset and extract the series to binarize
    dataset_copy = dataset.copy()
    X = dataset_copy[[column]]
    
    # Binarize the series
    binarizer = preprocessing.Binarizer(threshold=threshold).fit(X) 
    binarized_x = binarizer.transform(X)
    
    # Flatten the binarized field and return mutated copy
    dataset_copy[column] = pd.Series(binarized_x.reshape(-1))
    return dataset_copy
```

```python
b_df = binarize(df, 'thalach', 150)
b_df.head()
```

## Standardize data

- Standardization is a useful technique to transform attributes with a Gaussian distribution and differing means and standard deviations to a standard Gaussian distribution with a **mean of 0** and a **standard deviation of 1**.

```python
def standardize(dataset, column):
    # Copy dataset and extract the series to standardize
    dataset_copy = dataset.copy()
    X = dataset_copy[[column]]
    
    # Standardize the series
    standardizer = preprocessing.StandardScaler().fit(X) 
    standardized_x = standardizer.transform(X)
    
    # Flatten the binarized field and return mutated copy
    dataset_copy[column] = pd.Series(standardized_x.reshape(-1))
    return dataset_copy
```

```python
s_df = standardize(df, 'thalach')
s_df.head()
```
